```python
import numpy
import matplotlib.pyplot as plt
```

```python
# number of sample
num = [1, 10, 50, 100]
#list of sample means
means = []
```

```python
for j in num:
  numpy.random.seed(1)
  x = [numpy.mean(numpy.random.randint(-40, 40, j)) for i in range(j)]
  means.append(x)
```

```python
k=0#plotting from the 1st index
#plot the graphs
fig, ax = plt.subplots(2, 2, figsize=(6, 6))
for i in range(0, 2):
  for j in range(0, 2):
    # Histogram for each x stored in means
    ax[i, j].hist(means[k], 10, density = True)
    ax[i, j].set_title(label = num[k])
    k = k +1
plt.show()
```
